```python
# Creating a dataframe(df1) by reading "IMDB_Rating" CSV file
```

```python
import pandas as pd
df1 = pd.read_csv("IMDB_Rating.csv")
df1.head()
```

```python
# Creating a dataframe by reading a raw IMDB "movies_metadata" csv file
```

```python
df2 = pd.read_csv("movies_metadata.csv", low_memory=False)
df2.head()
```

```python
# Preprocess the movie metadata file to remove JSON columns
```

```python
df2 = df2.drop(['belongs_to_collection', 'genres', 'production_companies', 'production_countries','spoken_languages'], axis = 1)
df2.head()
```

```python
#Creating a dataframe(df3) by merging the df1 with df2 to create a single file with MovieID as index
#Creating a new dataframe(df4) that contains only the average "user rating" with the "revenue" per rating
```

```python
df3 = pd.merge(df1,df2,on='imdb_id')
df3.head()
frames = [df3['vote_average'], df3['revenue']]
df4 = pd.concat(frames, axis=1)
df4 = df4.drop_duplicates(subset=['vote_average'],keep='first')
df4 = df4.sort_values('vote_average')
df4.head()
```

```python
#Writing all the dataframes to CSV file
```

```python
df2.to_csv("processed_movies_metadat.csv")
df3.to_csv("merged_data_set.csv")
df4.to_csv("revenue_rating.csv")
```
